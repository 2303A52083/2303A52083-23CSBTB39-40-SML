From the above data:

1)Read the data with pandas and find features and target variables

2)Normalize the data with min-max scaling

3)Split the data into train and test.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
url1 = "https://www.kaggle.com/datasets/rohankayan/years-of-experience-and-salary-dataset"
data1 = pd.read_csv("Salary_Data.csv")
print(data1.head())
features1 = data1[['YearsExperience']]
target1 = data1['Salary']
print("Features:\n", features1.head())
print("Target:\n", target1.head())
plt.scatter(features1, target1, color='blue')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
scaler = MinMaxScaler()
normalized_data1 = scaler.fit_transform(data1)
normalized_df1 = pd.DataFrame(normalized_data1, columns=data1.columns)
print("Normalized Data:\n", normalized_df1.head())
```

Download the dataset from the above link.

1) Read the data with pandas and describe the data

2) Find data type and shape of each column

3) Find the target and features

4) Find the null values (if yes fill the null values with ‘0’ or mean of that column)

5) Normalize all the features

6) Split the data into train and test.



```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
data2 = pd.read_csv("diabetes_data_upload.csv")
print("First few rows of the dataset:\n", data2.head())
features2 = data2.drop('class', axis=1)
target2 = data2['class']
print("\nFeatures:\n", features2.head())
print("\nTarget:\n", target2.head())
label_encoder = LabelEncoder()
for column in features2.columns:
    if features2[column].dtype == 'object':
        features2[column] = label_encoder.fit_transform(features2[column])
plt.figure(figsize=(15, 10))
for i, column in enumerate(features2.columns[:6]):
    plt.subplot(2, 3, i + 1)
    plt.scatter(features2[column], label_encoder.fit_transform(target2))
    plt.title(f'{column} vs Target')
    plt.xlabel(column)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
scaler = MinMaxScaler()
normalized_features2 = scaler.fit_transform(features2)
normalized_df2 = pd.DataFrame(normalized_features2, columns=features2.columns)
print("\nNormalized Data:\n", normalized_df2.head())
```
