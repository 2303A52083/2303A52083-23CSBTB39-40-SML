1,:Given the following dataset: X={2.3,2.5,3.6,2.8,3.1,2.9,3.2,2.7,2.8,3.0}. Compute the density of the data using a histogram with 5 bins.From the above data .

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given dataset
data = [2.3, 2.5, 3.6, 2.8, 3.1, 2.9, 3.2, 2.7, 2.8, 3.0]

# Converting data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Computing histogram
# np.histogram returns the counts and bin edges
counts, bin_edges = np.histogram(df['Value'], bins=5, density=True)

# Printing histogram details
print("Counts (density):", counts)
print("Bin edges:", bin_edges)

# Ploting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Value'], bins=5, density=True, alpha=0.7, color='pink', edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Given Data with Density')

# Showing the plot
plt.grid(True)
plt.show()

# Convert data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Compute basic statistics
mean = df['Value'].mean()
median = df['Value'].median()
mode = df['Value'].mode().values
variance = df['Value'].var()
std_dev = df['Value'].std()
data_range = df['Value'].max() - df['Value'].min()
iqr = df['Value'].quantile(0.75) - df['Value'].quantile(0.25)

# Compute skewness and kurtosis
skewness = stats.skew(df['Value'])
kurtosis = stats.kurtosis(df['Value'])
```

Find Mean, Median, Mode , Variance , Standard deviation , Range , Interquartile Range (IQR),  Skewness ,Kurtosis of x and y .

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given data
x = [5, 6, 3, 8, 3, 7, 8]
y = [4, 5, 8, 2, 4, 2, 5]

# Create DataFrame
df = pd.DataFrame({
    'Independent Variable (x)': x,
    'Dependent Variable (y)': y
})

# Function to calculate and return statistics
def calculate_statistics(series):
    return {
        'Mean': series.mean(),
        'Median': series.median(),
        'Mode': series.mode().values,
        'Variance': series.var(),
        'Standard Deviation': series.std(),
        'Range': series.max() - series.min(),
        'Interquartile Range (IQR)': series.quantile(0.75) - series.quantile(0.25),
        'Skewness': stats.skew(series),
        'Kurtosis': stats.kurtosis(series)
    }

# Calculate statistics for x and y
x_stats = calculate_statistics(df['Independent Variable (x)'])
y_stats = calculate_statistics(df['Dependent Variable (y)'])

# Print statistics
print("Statistics for Independent Variable (x):")
for key, value in x_stats.items():
    print(f"{key}: {value}")

print("\nStatistics for Dependent Variable (y):")
for key, value in y_stats.items():
    print(f"{key}: {value}")

# Plotting
plt.figure(figsize=(14, 6))

# Histogram for x
plt.subplot(1, 2, 1)
plt.hist(df['Independent Variable (x)'], bins=5, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Independent Variable (x)')
plt.ylabel('Frequency')
plt.title('Histogram of x')

# Histogram for y
plt.subplot(1, 2, 2)
plt.hist(df['Dependent Variable (y)'], bins=5, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('Dependent Variable (y)')
plt.ylabel('Frequency')
plt.title('Histogram of y')

plt.tight_layout()
plt.show()

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Independent Variable (x)'], df['Dependent Variable (y)'], color='red')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Scatter Plot of x vs y')
plt.grid(True)
plt.show()
```
