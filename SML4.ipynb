"Question 1:

https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification?select=train.csvLinks to an external site.

From the above data:

1)Read the data with pandas and find features and target variables

2)Normalize the data with min-max scaling

3)Split the data into train and test.

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv('train.csv')
print(data.head())
features = data.drop('price_range', axis=1)
target = data['price_range']
print("Features:\n", features.head())
print("Target:\n", target.head())
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)
print("Scaled Features:\n", scaled_features_df.head())
X_train, X_test, y_train, y_test = train_test_split(scaled_features_df,target,test_size=0.2,random_state=42)
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

Download the dataset from the above link.

1) Read the data with pandas and describe the data

2) Find data type and shape of each column

3) Find the target and features

4) Find the null values (if yes fill the null values with ‘0’ or mean of that column)

5) Normalize all the features

6) Split the data into train and test.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv('train.csv')
print("Data Description:\n", data.describe())
print("\nData Types:\n", data.dtypes)
print("\nData Shape (Rows, Columns):", data.shape)
target = data['price_range']
features = data.drop('price_range', axis=1)
print("\nTarget Variable:\n", target.head())
print("\nFeatures:\n", features.head())
null_values = data.isnull().sum()
print("\nNull Values in each column:\n", null_values)
data_filled = data.fillna(0)
print("\nData after filling null values:\n", data_filled.head())
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)
print("\nScaled Features:\n", scaled_features_df.head())
X_train, X_test, y_train, y_test = train_test_split(scaled_features_df, target, test_size=0.2, random_state=42)
print("\nTraining data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```
